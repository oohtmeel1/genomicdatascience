```python
import re

def find_longest_orf(dna_sequence, index):
    # Find the starting and ending codons
    start = dna_sequence.find('ATG', index)
    longest_orf = ''
    longest_orf_length = 0
    while start != -1:
        end = dna_sequence.find('TAG|TAA|TGA', start)
        if end != -1:
            orf = dna_sequence[start:end+3]
            orf_length = len(orf)
            if orf_length > longest_orf_length:
                longest_orf = orf
                longest_orf_length = orf_length
        start = dna_sequence.find('ATG', end)

    # Return the longest ORF
    return longest_orf

# Open the FASTA file
with open('C:\gitfiles\Projects\Bioinformatics\dna2.fasta', 'r') as f:
    # Read the entire file contents
    contents = f.read()

# Use a regular expression to extract the DNA sequence
match = re.search(r'^[^>]+', contents, re.MULTILINE)
if match:
    dna = match.group(0)
else:
    dna = ''

# Initialize variables to keep track of the longest ORF
longest_orf = ''
longest_orf_length = 0

# Break the DNA sequence into overlapping fragments
fragment_length = len(dna) - 2
for i in range(0, len(dna), fragment_length):
    fragment = dna[i:i+fragment_length]
    orf = find_longest_orf(fragment, i)
    orf_length = len(orf)
    if orf_length > longest_orf_length:
        longest_orf = orf
        longest_orf_length = orf_length

# Print the longest ORF
print(longest_orf)
```
